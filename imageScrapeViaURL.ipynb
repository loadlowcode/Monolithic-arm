```python


import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin, urlparse
from zipfile import ZipFile
from google.colab import files

def download_images(url, folder_path):
    os.makedirs(folder_path, exist_ok=True)
    zip_filename = os.path.join(folder_path, 'images.zip')

    with ZipFile(zip_filename, 'w') as zipf:
        for img_url in (urljoin(url, img['src']) for img in BeautifulSoup(requests.get(url).text, 'html.parser').find_all('img', src=True)):
            img_name = os.path.basename(urlparse(img_url).path)
            img_path = os.path.join(folder_path, img_name)

            with open(img_path, 'wb') as img_file:
                img_file.write(requests.get(img_url).content)

            zipf.write(img_path, os.path.relpath(img_path, folder_path))

    return zip_filename

if __name__ == '__main__':
    zip_filename = download_images(input("Enter the URL to scrape images from: "), '/content/scraped_images')
    files.download(zip_filename)
```
