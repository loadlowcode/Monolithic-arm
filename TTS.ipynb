```python
from gtts import gTTS
from IPython.display import Audio, display
from google.colab import files

# Prompt the user to upload a .txt file
print("Please upload a .txt file.")
uploaded = files.upload()

# Check if a file is uploaded
if len(uploaded) > 0:
    # Get the filename
    file_name = list(uploaded.keys())[0]

    # Check if the file has a .txt extension
    if file_name.lower().endswith('.txt'):
        # Read the content of the uploaded file
        with open(file_name, 'r') as file:
            text_to_speak = file.read()

        # Specify the language (optional, default is English)
        language = 'en'

        # Create a gTTS object
        tts = gTTS(text=text_to_speak, lang=language, slow=False)

        # Save the speech as an audio file
        audio_file = 'output.mp3'
        tts.save(audio_file)

        # Display the audio file
        display(Audio(audio_file))
    else:
        print("Invalid file format. Please upload a .txt file.")
else:
    print("No file uploaded.")
```
